# Model Building & Finding out the Accuracy

```python
# Importing of Libraries
import pandas as pd
import numpy as np
```

```python


dataset=pd.read_csv('Enhanced_Mining_dataset.csv',sep=',')
dataset=dataset.set_index('date')
dataset.head()
```

```python
#dataset=dataset.set_index('date')
dataset.head()
```

```python
x=dataset.iloc[:,:-1].values

#Dependent Variable
y=dataset.iloc[:,-1].values
```

```python
# Independent Variables
x
```

```python
# Dependent Variable
y
```

```python
# Spliting of Dataset into Trainset & Testset

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Importing the Random forest regressor model
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=42,criterion='mse')
# Fitting the Random forest regressor to x_train & y_train
model.fit(x_train,y_train)
```

```python

# Prediction

y_pred=model.predict(x_test)
y_pred
```

```python
# Finding Accuracy of the model
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(model,open('mining.pkl','wb'))
```
